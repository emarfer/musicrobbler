```python
import pandas as pd
import os
import numpy as np
import sqlalchemy as alch
from datetime import datetime as dt
import dotenv
dotenv.load_dotenv()
```

```python
# conection to mysql
passw = os.getenv("mysql")
dbName = "musicrobbler"
connectionData = f"mysql+pymysql://root:{passw}@localhost/{dbName}"
engine = alch.create_engine(connectionData)
```

```python
# read table temporadas
temporadas = pd.read_sql_query(f'''
select * from temporadas
''',engine)
temporadas.head()
```

```python
# split name temporadas
new_cols = [1,2,3,4]
temporadas[new_cols] = temporadas.temporada.str.split('_',expand=True)
```

```python
temporadas.head()
```

```python
# changing fields elements
```

```python
temporadas.loc[temporadas[3] == 'fem',4] ='fem'
temporadas.loc[temporadas[3] == 'fem',3] = np.nan
```

```python
temporadas.loc[temporadas[1] == 'anual',3] = 'AA'
temporadas.loc[temporadas[1] == 'cumple',3] = 'CC'
```

```python
temporadas['order'] = temporadas[3]
```

```python
# meses dias
# expand fechas
temporadas.info()
```

```python
temporadas['inicio_year'] = pd.DatetimeIndex(temporadas.inicio).year
temporadas['inicio_month'] = pd.DatetimeIndex(temporadas.inicio).month
temporadas['inicio_day'] = pd.DatetimeIndex(temporadas.inicio).day

temporadas['fin_year'] = pd.DatetimeIndex(temporadas.fin).year
temporadas['fin_month'] = pd.DatetimeIndex(temporadas.fin).month
temporadas['fin_day'] = pd.DatetimeIndex(temporadas.fin).day

```

```python
temporadas.head()
```

```python
temporadas.info()
```

```python
temporadas.loc[temporadas[4].isna(),'temporada_name'] = temporadas[2].astype('str') + '_' + temporadas.order  
temporadas.loc[temporadas[4].isna() == False,'temporada_name'] = temporadas[2].astype('str') + '_' + temporadas.order  +'_fem'
```
